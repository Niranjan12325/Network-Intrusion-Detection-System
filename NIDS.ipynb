```python
# ==============================
# Fast AI-Powered Intrusion Detection System (IDS)
# Sample Dataset included
# Algorithms: Decision Tree & Random Forest
# ==============================

# --- 1. Import libraries ---
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# -----------------------------
# 2. Load Sample Dataset
# -----------------------------
data_path = '/content/sample_nids.csv'  # Make sure this file is in the same folder
data = pd.read_csv(data_path)

# For simplicity, use all data as both train and test
train_data = data.copy()
test_data  = data.copy()

# -----------------------------
# 3. Encode categorical features
# -----------------------------
for col in ["protocol_type", "service", "flag"]:
    le = LabelEncoder()
    train_data[col] = le.fit_transform(train_data[col])
    test_data[col]  = le.transform(test_data[col])

# Encode labels: 0 = normal, 1 = attack
train_data['label'] = train_data['label'].apply(lambda x: 0 if x=='normal' else 1)
test_data['label']  = test_data['label'].apply(lambda x: 0 if x=='normal' else 1)

# -----------------------------
# 4. Split features and labels
# -----------------------------
X_train = train_data.drop(["label"], axis=1)
y_train = train_data["label"]
X_test  = test_data.drop(["label"], axis=1)
y_test  = test_data["label"]

# -----------------------------
# 5. Feature scaling
# -----------------------------
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)

# ==============================
# 6. Train Decision Tree
# ==============================
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)

# ==============================
# 7. Train Random Forest
# ==============================
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# ==============================
# 8. Evaluation
# ==============================
print("Decision Tree Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_dt)*100))
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_dt))

print("\nRandom Forest Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_rf)*100))
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))

# ==============================
# 9. Print Normal or Attack for each record
# ==============================
print("\nDecision Tree Predictions:")
for i, pred in enumerate(y_pred_dt):
    print(f"Record {i+1}: {'Normal' if pred==0 else 'Attack'}")

print("\nRandom Forest Predictions:")
for i, pred in enumerate(y_pred_rf):
    print(f"Record {i+1}: {'Normal' if pred==0 else 'Attack'}")
```
